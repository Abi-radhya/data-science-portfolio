```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Cardiotocographic.csv')

# View first few rows
df.head()

# Check for missing values
df.isnull().sum()

# Check data types
df.dtypes

# Describe the data before treating outliers
df.describe()
```

```python
# Central tendency and dispersion
summary = df.describe().T
summary['median'] = df.median()
summary['IQR'] = df.quantile(0.75) - df.quantile(0.25)
summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms
df.hist(figsize=(15, 10), bins=15)
plt.tight_layout()
plt.show()

# Boxplots
plt.figure(figsize=(15, 8))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()

# Heatmap for correlation
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Pairplot for selected variables
sns.pairplot(df[['LB', 'AC', 'FM', 'UC', 'ASTV']])
plt.show()

# Violin plot example
plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x='UC', y='LB')
plt.title("Fetal Heart Rate by Uterine Contractions")
plt.show()
```

```python
# 4. Pattern Recognition and Insights
# LB (Baseline Heart Rate) shows a moderate positive correlation with UC (Uterine Contractions), and a negative correlation with DL and DP, indicating possible fetal distress when decelerations increase.

# ASTV and MSTV are strongly correlated, showing aligned variability patterns.

# FM (Fetal Movements) has a slight positive link to LB, suggesting healthier fetal activity.

# No time-based data is present, so temporal trends can't be assessed.
```
