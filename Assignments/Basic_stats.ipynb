```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("sales_data_with_discounts.csv")  
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

sns.histplot(data['Volume'], kde=True)
plt.title('Distribution of Volume', fontsize=20, color="hotpink")
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Avg Price'], kde=True)
plt.title('Distribution of Avg Price', fontsize=20, color="yellow")
plt.xlabel('Avg Price')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Total Sales Value'], kde=True)
plt.title('Distribution of Total Sales Value', fontsize=20, color="orange")
plt.xlabel('Total Sales Value')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Discount Rate (%)'], kde=True)
plt.title('Distribution of Discount Rate', fontsize=20, color="red")
plt.xlabel('Discount Rate (%)')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Discount Amount'], kde=True)
plt.title('Distribution of Discount Amount', fontsize=20, color="green")
plt.xlabel('Discount Amount')
plt.ylabel('Frequency')
plt.show()

sns.histplot(data['Net Sales Value'], kde=True)
plt.title('Distribution of Net Sales Value', fontsize=20, color="purple")
plt.xlabel('Net Sales Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histograms for the numerical columns show how values are distributed. 
# For example, some columns like Avg Price and Total Sales Value are right-skewed, indicating a few very high values. 
# Columns like Volume are more normally distributed. This helps identify skewness and potential outliers.
```

```python
# Standardization, also known as z-score normalization, is a technique used to bring all numerical variables onto a common scale 
# without distorting differences in the ranges of values. 
# It involves transforming the data so that it has a mean of 0 and a standard deviation of 1. 
# This is done by subtracting the mean of a column from each value and then dividing the result by the standard deviation of that column. 
# The formula used is: z = (x - Î¼) / Ïƒ, where x is the original value, Î¼ is the mean, and Ïƒ is the standard deviation. 
# Standardizing data is especially useful when comparing features with different units or scales, 
# and it helps improve the performance and accuracy of many analytical and machine learning models.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

for col in num_cols:
    plt.figure(figsize=(5, 3))
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'{col} - Before Standardization')
    plt.tight_layout()
    plt.show()

for col in num_cols:
    mean = data[col].mean()
    std = data[col].std()
    data[col + '_std'] = (data[col] - mean) / std

for col in num_cols:
    plt.figure(figsize=(5, 3))
    sns.histplot(data[col + '_std'], kde=True, color='lightgreen')
    plt.title(f'{col} - After Standardization')
    plt.tight_layout()
    plt.show()
```
